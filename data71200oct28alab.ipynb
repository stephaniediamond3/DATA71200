# Import Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Import Wine Dataset

```python
from sklearn.datasets import load_wine
wine_dataset = load_wine()
```

# Testing/Training Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    wine_dataset.data, wine_dataset.target, stratify=wine_dataset.target, random_state=66,test_size=0.2)
```

# Import Logistic Regression and Perform Grid Search

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# parameter validation on training set

#dictionary of all values of C
param_grid = {"C": [0.001, 0.1, 1, 10]}

#grid search on all values of C in dictionary
logreg_gscv = GridSearchCV(LogisticRegression(), param_grid, cv=5)
logreg_gscv.fit(X_train, y_train)
```

```python
#try increasing max_iter
logreg_gscv = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
logreg_gscv.fit(X_train, y_train)
```

```python
#dictionary of all values of c
param_grid = {"C": [0.001, 0.1, 1, 10]}

#grid search on all values of k in dictionary for logist regression
logreg_gscv = GridSearchCV((logreg_gscv), param_grid, cv=5)
logreg_gscv.fit(X_train, y_train)
```

```python
#best performing c (on training set)
print("Best C on training set:", logreg_gscv.best_params_)
```
